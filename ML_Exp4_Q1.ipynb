```python
df1=df[["radius_mean","perimeter_mean","smoothness_mean","symmetry_mean","smoothness_se","radius_worst","smoothness_worst","symmetry_worst","symmetry_se","diagnosis"]]
df1.head()
```

```python
def calculate_prior(df1, Y):
  classes = sorted(list(df1[Y].unique()))
  prior = []
  for i in classes:
    prior.append(len(df1[df1[Y]==i])/len(df1))
  return prior
```

```python
def calculate_likelihood_gaussian(df1, feat_name, feat_val, Y, label):
  feat = list(df1.columns)
  df1 = df1[df1[Y]==label]
  mean, std = df1[feat_name].mean(), df1[feat_name].std()
  p_x_given_y = (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-((feat_val-mean)**2 / (2 * std**2 )))
  # p_x_given_y = len(df1[df1[feat_name]==feat_val]) / len(df)
  return p_x_given_y
```

```python
def naive_bayes_gaussian(df1, X, Y):
  features = list(df1.columns)[:-1]
  prior = calculate_prior(df1, Y)
  Y_pred = []
  for x in X:
    labels = sorted(list(df1[Y].unique()))
    likelihood = [1]*len(labels)
    for j in range(len(labels)):
      for i in range(len(features)):
        likelihood[j] *= calculate_likelihood_gaussian(df1, features[i], x[i], Y, labels[j])
      post_prob = [1]*len(labels)
    for j in range(len(labels)):
      post_prob[j] = likelihood[j] * prior[j]
    Y_pred.append(np.argmax(post_prob))
  return np.array(Y_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score
```

```python

train, test = train_test_split(df1, test_size=0.25, random_state=41)
X_test = test.iloc[:,:-1].values
Y_test = test.iloc[:,-1].values
Y_pred = naive_bayes_gaussian(train, X=X_test, Y="diagnosis")
Y_pred = map(str,Y_pred)
Y_pred = list(Y_pred)
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
print(f1_score(Y_test,Y_pred,pos_label='1'))
```
