```python
#import
import pandas as pd
import numpy as np
```

```python
#Load the dataset
df=pd.read_csv('/content/Social_Network_Ads.csv')
df
```

```python
x=df.iloc[:,[1,2,3]].values # Selecting First,Second and Third Column
y=df.iloc[:,-1].values # Selecting Last Column
```

```python
x # Data in original Male Female form
```

```python
#Preprocessing
#Converting characters into numeric
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:,0]=le.fit_transform(x[:,0])
df
```

```python
x # data  in 0,1 form
```

```python
#spliting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Feature Scaling/ Normalization(To convert Unstructured Data into Structured )
from sklearn.preprocessing import StandardScaler # StandardScaler use for normailising feature value
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Fit and Predict
from sklearn.naive_bayes import GaussianNB # GaussianNB -- Library
classifier=GaussianNB() # classifier-- object of GaussianNB Function
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
y_pred
```

```python
# Evaluation
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,y_pred)
ad=accuracy_score(y_test,y_pred)
cm # Printing Confusion Matrix
```

```python
ad # Printing Accuracy Score
```

2. Bayesian Classifier Using German Credit DataSet

```python
#Load the dataset
df=pd.read_csv('/content/german_credit_data (1).csv')
df
```

```python
x=df.iloc[:,[1,2,3]].values # Selecting Second, Third and Fourth Column
y=df.iloc[:,-1].values # Selecting Last Column
```

```python
x # Data in original Male Female form
```

```python
#Preprocessing
#Converting characters into numeric
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:,0]=le.fit_transform(x[:,0])
df
```

```python
x # data in 0,1
```

```python
#spliting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling/ Normalization(To convert Unstructured Data into Structured )
from sklearn.preprocessing import StandardScaler # StandardScaler use for normailising feature value
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Fit and Predict
from sklearn.naive_bayes import GaussianNB # GaussianNB -- Library
classifier=GaussianNB()  # classifier-- object of GaussianNB Function
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
y_pred
```

```python
# Evaluation
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,y_pred)
ad=accuracy_score(y_test,y_pred)
cm # Printing Confusion Matrix
```
