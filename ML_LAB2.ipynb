```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[0.5,2.5]
y=[0.2,0.9]
e=[]
weight=[]
bias=[]
```

```python
def dw(x,y,w,b,c):
  yin=x*w+b
  ynot=sig(yin)
  fdash=ynot*(1-ynot)
  return c*(y-ynot)*fdash*x
```

```python
def db(x,y,w,b,c):
  yin=x*w+b
  ynot=sig(yin)
  fdash=ynot*(1-ynot)
  return c*(y-ynot)*fdash
```

```python
def error(x,y,w,b):
  yin=x*w+b
  ynot=sig(yin)
  err=(y-ynot)**2
  return err
```

```python
def sig(x): #sigmoid func
 return 1/(1 + np.exp(-x))
```

```python
def stochastic(x,y,w,b):
  wt=w;
  bi=b;
  c=1;epoch=1000;
  for j in range(epoch):
    for i in range(len(x)):
      wnew=wt+dw(x[i],y[i],wt,bi,c)
      bnew=bi+db(x[i],y[i],wt,bi,c)
      err=error(x[i],y[i],wt,bi)
      e.append(err)
      weight.append(wnew)
      bias.append(bnew)
      wt=wnew
      bi=bnew
      print(wnew)
      print(bnew)
```

```python
stochastic(x,y,-2,-2)
```

```python
xpoints = np.array(weight)
ypoints = np.array(e)
plt.ylabel("error")
plt.xlabel("weight")
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array(bias)
ypoints = np.array(e)
plt.ylabel("error")
plt.xlabel("bias")
plt.plot(xpoints, ypoints)
plt.show()
```
